```python
# redimention de tensors

tensor = torch.randn(4, 4)
nouveau_tensor = tensor.view(16)
print(tensor.shape) # torch.Size([4, 4])
print(tensor)
print(nouveau_tensor.shape) # torch.Size([16])
print(nouveau_tensor)
```

```python
tensor = torch.randn(2, 3, 4)
print(tensor)
nouveau_tensor = tensor.reshape(4, -1)
print(nouveau_tensor)
print(tensor.shape) # torch.Size([2, 3, 4])
print(nouveau_tensor.shape) # torch.Size([6, 4]))
```
